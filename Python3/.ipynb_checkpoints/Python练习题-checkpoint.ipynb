### 输出1+100的结果

```python
sum = 0
for i in range(100):
    sum += i+1
print(sum)
```

```python
sum = 0
l = 1
r = 100
sum = 0
while(l<r):
    sum += l+r
    l += 1
    r -= 1 
sum
```

###  打印100以内的质数

```python
def judge(x):
    #因为range(2,2)是空的，取不到2，直接跳过循环里的判断
    for i in range(2,x):
        if x % i == 0:
            return False
    return True

for i in range(2,101):
    if judge(i):
        print(i)
    else:
        continue
```

### 计算一个文件中每个英文单词出现的次数

```python
f = open('test.txt')
lines = f.readlines()
#定义一个字典，放单词和出现的次数
count = {}
for line in lines:
    #单词都是以空格结尾的，行是以/n结尾的
    tokens = line.strip().split(' ')
    for token in tokens:
        #先判断这个单词有没有在字典中出现过
        if token in count:
            count[token] += 1
        #没有的话在字典里创建这个key并初始化为1
        else:
            count[token] = 1

for word in count:
    print(word,count[word])
            
    
```
